# Four example circuits
To simulate the classification using PyTorch, we decided to pick four circuits from Hannah's paper
https://arxiv.org/pdf/1905.10876.pdf

Circuits are chosen to explore different expressibility and entangling capabilities.

By looking at the figure I have decided to use circuit9,circuit1,circuit14,circuit6 because those circuits can capture 4 different expressibility and entangling capability

Expressibility

9 (Very Low) 1 (low) 14(high) 6 (Very high)

Entalging capbility

1 (Very Low) 14(low) 6(high) 9(very high)

```python
import sys
sys.path.append('../../Pyfiles')
import qiskit
from circuits import *
```

```python
from qiskit.circuit import Parameter

#circuit 9
qr = QuantumRegister(4)
qc = QuantumCircuit(qr)

theta=[]
for i in range(20):
    theta.append((Parameter('θ'+str(i))))
qc=circuit9(qc,qr,theta,1,0)
qc.draw('mpl')
```
