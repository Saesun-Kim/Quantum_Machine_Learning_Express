# Dataset Generation 
I am following from the scikit-learn website and blog(machinelearningmastery.com) to generate the dataset

```python
# example of binary classification task
from numpy import where
from collections import Counter
from sklearn.datasets import make_blobs
from matplotlib import pyplot

# define dataset
centers = [(0.35, 0.25), (0.65, 0.25)]

X, y = make_blobs(n_samples=1500, centers=centers,cluster_std=0.08, random_state=1)
# summarize dataset shape

X[:,1:] = X[:,1:] * 2
print(X.shape, y.shape)
# summarize observations by class label
counter = Counter(y)
print(counter)
# summarize first few examples
for i in range(10):
	print(X[i], y[i])
# plot the dataset and color the by class label
for label, _ in counter.items():
	row_ix = where(y == label)[0]
	pyplot.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label))
pyplot.legend()
pyplot.xlim([0, 1])
pyplot.ylim([0, 1])
pyplot.show()
```
